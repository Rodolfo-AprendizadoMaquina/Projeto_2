# Análise de Componentes Principais (PCA)

```python
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import svd
```

## Visualização dos Exemplos de Treinamento

```python
mat = loadmat("dado3.mat")
X = mat["X"]
```

```python
rnd_ind = np.random.choice(X.shape[0], size=100, replace=False)
print_images = []
for i in rnd_ind:
    print_images.append(X[i,:])
```

```python
fig, axis = plt.subplots(10, 10,figsize=(8,8))
for i in range(10):
    for j in range(10):
        axis[i,j].imshow(print_images[10*i+j].reshape(32,32,order="F"),cmap='gray')
        axis[i,j].axis("off")
plt.show()
```

## Determinação e visualização das *eigenfaces*

```python
Sigma = 1/X.shape[0] * (X.T @ X)
U, S, V = svd(Sigma)
```

```python
fig, axis = plt.subplots(6, 6,figsize=(8,8))
for i in range(6):
    for j in range(6):
        axis[i,j].imshow(U[:,6*i+j].reshape(32,32,order="F"),cmap='gray')
        axis[i,j].axis("off")
plt.show()
```

## Projeção das faces

```python
n_comp = 100
U_red = U[:,:n_comp]
z = X @ U_red
```

```python
proj = z @ U_red.T
print_proj = []
for i in rnd_ind:
    print_proj.append(proj[i,:])
```

```python
fig, axis = plt.subplots(10, 10,figsize=(8,8))
for i in range(10):
    for j in range(10):
        if np.mod(i,2) == 0:
            axis[i,j].imshow(print_images[5*i+j].reshape(32,32,order="F"),cmap='gray')
            axis[i,j].axis("off")
        else:
            axis[i,j].imshow(print_proj[5*(i-1)+j].reshape(32,32,order="F"),cmap='gray')
            axis[i,j].axis("off")
plt.show()
```

```python
fig, axis = plt.subplots(10, 10,figsize=(8,8))
for i in range(10):
    for j in range(10):
        if np.mod(i,2) == 0:
            axis[i,j].imshow(print_images[50+5*i+j].reshape(32,32,order="F"),cmap='gray')
            axis[i,j].axis("off")
        else:
            axis[i,j].imshow(print_proj[50+5*(i-1)+j].reshape(32,32,order="F"),cmap='gray')
            axis[i,j].axis("off")
plt.show()
```

## Fração da variância original mantida

```python
plot_var = []
for i in range(50, X.shape[1]):
    plot_var.append(np.sum(S[:i]) / np.sum(S))
    
plt.plot(np.arange(50,X.shape[1]), plot_var)
plt.xticks(np.arange(50,X.shape[1],100))
plt.yticks(np.arange(0.89,1.01,0.01))
plt.xlabel('Número de componentes principais')
plt.ylabel('Fração da variância original mantida')
plt.grid()
plt.title('Fração da variância  com relação ao n° de componentes principais')
plt.show()
```
